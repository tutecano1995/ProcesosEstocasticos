### Ejercicio 1 - Cálculo de ganancia de los repetidores analógicos

> _Asuma que cada repetidor analógico puede transmitir como máximo con una energía promedio $\varepsilon$. Determine el valor de A y las ganancias $G_{2}, . . . ,G_{n}$ para satisfacer ese requisito. Exprese las ganancias en función de la relación señal a ruido a la salida de cada repetidor:_

> $$SNR = h^{2} \varepsilon / \sigma^{2}$$

[comment]: <> (&nbsp; agrega un espacio)

En primer lugar, sabemos que la energía promedio del simbolo transmitido para &nbsp;$X_{i}$ es:

$$\varepsilon_{X_{i}} = {\sigma^{2}}_{X_{i}} = \mathbb{E}[{X_{i}}^2] - \mathbb{E}[{X_{i}}]^{2}$$

Analicemos entonces los valores que toma $\mathbb{E}[{X_{i}}]^{2}$, sabiendo que:

$$X_{i} = G_{i}\;(h\; X_{i-1}+W_{i})$$

Podemos decir que:

$$ \mathbb{E}[X_{i}] = \mathbb{E}[G_{i}\;(h\; X_{i-1}+W_{i})] = G_{i}\;(h\;\mathbb{E}[X_{i-1}]+\mathbb{E}[W_{i}])$$

*Nota: tanto $h$ como $G_{i}$ son constantes*

como:

$$\mathbb{E}[X_{1}] = 0$$

$$\mathbb{E}[W_{i}] = 0$$

Entonces:

$$ \mathbb{E}[X_{i}] = 0 $$

Por lo tanto:

$$\varepsilon_{X_{i}} = {\sigma^{2}}_{X_{i}} = \mathbb{E}[{X_{i}}^2]$$

Ahora veamos cuanto vale $\mathbb{E}[{X_{i}}^2]$:

$$\mathbb{E}[{X_{i}}^2] 
 = \mathbb{E}[{G_{i}}^{2}\;{(h\; X_{i-1}+W_{i})}^{2}] 
 = {G_{i}}^{2}\;(\mathbb{E}[{X_{i-1}}^2] + 2\;h\;\mathbb{E}[X_{i-1}\;W_{i-1}] + \mathbb{E}[{W_{i-1}}^2])$$

como:

$$X_{i}\perp\!\!\!\perp W_{i}$$

Entonces:

$$\mathbb{E}[X_{i-1}\;W_{i-1}] = \mathbb{E}[X_{i-1}] \; \mathbb{E}[W_{i-1}] = 0$$

Y reemplazando $\mathbb{E}[{X_{i}}^2]$ por $\varepsilon_{X_{i}}$ y $\mathbb{E}[W_{i}]$ por ${\sigma}^{2}$:

$$\varepsilon_{X_{i}} = {G_{i}}^{2}\;(\varepsilon_{X_{i-1}\;{h}^{2}}  + {\sigma}^{2})$$

Habiendo obtenido esta expresión, y sabiendo que se busca que se transmita como máximo con una energía de $\varepsilon$, podemos plantear que todos los repetidores tienen la energía máxima. Es decir:

$$\varepsilon_{X_{i}} = \varepsilon \quad\forall i \in \mathbb{N}$$

Entonces:

$$\varepsilon_{X_{1}} = {A}^{2} = \varepsilon$$

$$\varepsilon_{X_{i}} = {G_{i}}^{2}\;(\varepsilon\;h + {\sigma}^{2}) = \varepsilon$$

$$ {G_{i}}^{2}
 = \frac{\varepsilon}{\varepsilon\;{h}^{2} + \sigma^{2}} 
 = \frac{\varepsilon}{\frac{\varepsilon\;{h}^{2}}{\sigma^{2}}\;\sigma^{2} + \sigma^{2}} 
 = \frac{\varepsilon}{(SNR+1)\;\sigma^{2}}
 = \frac{\varepsilon\;h^{2}}{(SNR+1)\;\sigma^{2}\;h^{2}}
 = \frac{SNR}{(SNR+1)\;h^{2}}$$


### Ejercicio 2 - Probabilidad de error del sistema analógico

> _1. Exprese $Y_{n}$ en función de $X_{1}$, las ganancias y los ruidos $W_{1},...,W_{n}$. ¿Que distribución tiene los términos asociados sólo al ruido?_

Primero veamos los valores que toma $Y_{n}$ para distintos valores de n (con n = 1, n = 2 y n = 3):

$$Y_{1} = h\;X_{1} + W_{1}$$

$$Y_{2} = h\;X_{2} + W_{2} = h\;(G_{2}\;Y_{1}) + W_{2} = h\;(G_{2}\;(h\;X_{1} + W_{1})) + W_{2} $$
$$Y_{2} = G_{2}\;h^{2}\;X_{1} + G_{2}\;h\;W_{1}+W_{2}$$

$$Y_{3} = h\;X_{3} + W_{3} = h\;(G_{3}\;Y_{2}) + W_{3} = h\;(G_{3}\;(h\;(G_{2}\;(h\;X_{1} + W_{1})) + W_{2})) + W_{3}$$
$$Y_{3} = G_{3}\;G_{2}\;h^{3}\;X_{1} + G_{3}\;G_{2}\;h^{2}\;W_{1} + G_{3}\;h\;W_{2} + W_{3}$$

Veamos ahora la expresión general:

$$Y_{n} = h\;X_{n} + W_{n} = h\;(G_{n}\;Y_{n-1}) + W_{n}$$

Desarrollando el término $Y_{n-1}$:

$$Y_{n} = h\;X_{n} + W_{n} = h\;(G_{n}\;(h\;(G_{n-1}\;Y_{n-2}) + W_{n-1})) + W_{n}$$

Y se podría continuar desarrollando los susecivos términos $Y_{n-k}$ hasta que $k = n-1$. Realizando esto se obtendría:

$$Y_{n} = (\prod^{n}_{i=2}G_{i})\;h^{n}\;X_{1} + \sum^{n}_{i=1}(\prod^{n}_{j=i+1}G_{j}\;h)\;W_{i}$$

Analizando esta ecuación, los terminos asociados al ruido son los que contienen $W_{i}$. Como $W_{i}\sim\mathcal{N}(0,\sigma^{2})$ y en cada término $W_{i}$ está multiplicado por un factor $\prod^{n}_{j=i+1}G_{j}\;h$ que es constante, entonces los términos también tienen distribución normal de $\mu = 0$. Incluso la suma de todos los términos también tendrá distribución normal de $\mu = 0$. 

Análogamente, el término restante (que contiene $X_{1}$ es el asociado a la señal que llega en la última etapa. Sabiendo que $X_{1}$ puede tomar los valores $A$ o $-A$ con probabilidad $0.5$, este término podrá valer  $(\prod^{n}_{i=2}G_{i})\;h^{n}\;A$ o $-(\prod^{n}_{i=2}G_{i})\;h^{n}\;A$ con probabilidad $0.5$ ambos.

Si se reemplazan los valores obtenidos de $G_{i}$ en $Y_{n}$:

$$Y_{n} = (\prod^{n}_{i=2}\;\sqrt{\frac{SNR}{SNR+1}}\;\frac{1}{h})\;h^{n}\;X_{1} + \sum^{n}_{i=1}(\prod^{n}_{j=i+1}\sqrt{\frac{SNR}{SNR+1}}\;\frac{1}{h}\;h)\;W_{i}$$

$$Y_{n} = (\prod^{n}_{i=2}\;\sqrt{\frac{SNR}{SNR+1}}\;\frac{1}{h})\;h^{n}\;X_{1} + \sum^{n}_{i=1}(\prod^{n}_{j=i+1}\sqrt{\frac{SNR}{SNR+1}})\;W_{i}$$

$$Y_{n} = (\sqrt{\frac{SNR}{SNR+1}})^{n-2}\;X_{1} + \sum^{n}_{i=1}(\sqrt{\frac{SNR}{SNR+1}})^{n-(i+1)}\;W_{i}$$

Teniendo en cuenta esta expresión se puede obtener la varianza del término asociado al ruido en $Y_{n}$. Cómo se expreso anteriormente, cada $W_{i}$ está multiplicado por un factor constante, por lo tanto:

$$Y_{n} = (\sqrt{\frac{SNR}{SNR+1}})^{n-2}\;X_{1} + \sum^{n}_{i=1}W_{i}^{*}$$

Donde $W_{i}^{*}\sim\mathcal{N}(0,(\sqrt{\frac{SNR}{SNR+1}})^{\frac{n-(i+1)}{2}} \;\sigma^{2})$.

Como $W_{i}$ son independientes, $W_{i}^{*}$ también lo son, y por lo tanto:

$$\sum^{n}_{i=1}W_{i}^{*}\sim \mathcal{N}(0,\sigma^{2}\;\sum^{n}_{i=1}(\sqrt{\frac{SNR}{SNR+1}})^{\frac{n-(i+1)}{2}} )$$


> _2. Calcule la relación señal a ruido en la última etapa $\rho_{n}$._

La relación señal a ruido en la última etápa $\rho_{n}$ la podemos definir el cociente entre la energía de la señal en $\rho_{n}$ y la varianza del ruido en $\rho_{n}$. Como del punto anterior se pudieron identificar los términos asociados a la señal y los asociados al ruido, podemos expresar:

$${SNR}_{\rho_{n}} = \frac{var[(\prod^{n}_{i=2}G_{i})\;h^{n}\;X_{1}]}{var[\sum^{n}_{i=1}(\prod^{n}_{j=i+1}G_{j}\;h)\;W_{i}]}$$

Sabiendo que $(\prod^{n}_{i=2}G_{i})\;h^{n}$ es constante:

$${SNR}_{\rho_{n}} = \frac{[(\prod^{n}_{i=2}G_{i})\;h^{n}]^{2}\;var[X_{1}]}{var[\sum^{n}_{i=1}(\prod^{n}_{j=i+1}G_{j}\;h)\;W_{i}]}$$

como:

$$W_{i}\perp\!\!\!\perp W_{j}\quad \forall\;i \neq j $$ 

y los factores $(\prod^{n}_{j=i+1}G_{j}\;h)^{2}$ son constantes, entonces:

$${SNR}_{\rho_{n}} = \frac{[(\prod^{n}_{i=2}G_{i})\;h^{n}]^{2}\;var[X_{1}]}{\sum^{n}_{i=1}\;(\prod^{n}_{j=i+1}G_{j}\;h)^{2}\; var[W_{i}]}$$


Y reemplazando $var[X_{1}]$ por $A^{2}$ y $var[X_{i}]$ por $\sigma^{2}$:

$${SNR}_{\rho_{n}} = \frac{[(\prod^{n}_{i=2}G_{i})\;h^{n}]^{2}\;A^{2}}{\sum^{n}_{i=1}\;(\prod^{n}_{j=i+1}G_{j}\;h)^{2}\; \sigma^{2}}$$

Finalmente, si reemplazamos los valores obtenidos de $G_{i}$:

$${SNR}_{\rho_{n}} = \frac{(\frac{SNR}{SNR+1})^{n-2}\;SNR}{\sum_{i=1}^{n+1}(\frac{SNR}{SNR+1})^{n-(i+1)}}$$


> _3. Calcule la probabilidad de error promedio en función de la SNR de la última etapa $\rho_{n}$._

En primer lugar planteamos la ecuación para determinar la probabilidad de error:

$$P_{e,n} = \frac{1}{2}(P_{e|X_{1}=A} + P_{e|X_{1}=-A})$$

Determinemos entonces los valores de $P_{e|X_{1}=A}$ y $P_{e|X_{1}=-A}$. Como sabemos que la última decisión se toma en base al valor de $Y_{n}$, podemos plantear que:

$$P_{e|X_{1}=A} = P(Y_{n}<0) = P((\prod^{n}_{i=2}G_{i})\;h^{n}\;A + \sum^{n}_{i=1}(\prod^{n}_{j=i+1}G_{j}\;h)\;W_{i}<0)$$

$$P_{e|X_{1}=A} = P(\sum^{n}_{i=1}(\prod^{n}_{j=i+1}G_{j}\;h)\;W_{i}<-(\prod^{n}_{i=2}G_{i})\;h^{n}\;A)$$

como:

$$W_{i}\perp\!\!\!\perp W_{j}\quad \forall\;i \neq j $$ 

y además $(\prod^{n}_{j=i+1}G_{j}\;h)$ es constante y $var(W_{i}) = \sigma^{2}$. Entonces:

$$P_{e|X_{1}=A} = P(\sqrt{\sum^{n}_{i=1}(\prod^{n}_{j=i+1}G_{j}\;h)\;\sigma^{2}}\;Z<-(\prod^{n}_{i=2}G_{i})\;h^{n}\;A)$$

Donde $Z\sim \mathcal{N}(0,1)$.

Como la distribución normal con $\mu=0$ es simétrica:

$$P_{e|X_{1}=A} = P(\sqrt{\sum^{n}_{i=1}(\prod^{n}_{j=i+1}G_{j}\;h)\;\sigma^{2}}\;Z>(\prod^{n}_{i=2}G_{i})\;h^{n}\;A)$$

$$P_{e|X_{1}=A} = P(Z>\frac{(\prod^{n}_{i=2}G_{i})\;h^{n}\;A)}{\sqrt{\sum^{n}_{i=1}(\prod^{n}_{j=i+1}G_{j}\;h)\;\sigma^{2}}})$$

Tomando en cuenta el resultado del punto anterior, se puede concluir:

$$P_{e|X_{1}=A} = P(Z>\sqrt{{SNR}_{\rho_{n}}}) = Q(\sqrt{{SNR}_{\rho_{n}}})$$

Realizando un desarrollo análogo, también se obtiene que:

$$P_{e|X_{1}=-A} = Q(\sqrt{{SNR}_{\rho_{n}}})$$

Por lo tanto:

$$P_{e,n} = Q(\sqrt{{SNR}_{\rho_{n}}})$$

> _Para el sistema de comunicaciones digital es posible demostrar la probabilidad de error es_

> 1. Grafique en una misma figura las probabilidades de error del sistema digital y del sistema analógico en función de la $SNR$, con $SNR \in [−5,30]$ dB (con pasos de 1 dB).
```python
import scipy.special
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

q_func = lambda x: 0.5-0.5*scipy.special.erf(x/np.sqrt(2))

#SNR en la última etapa
SNR_rho_n = lambda n, SNR: (float(SNR)/float(SNR+1))**(n-2)*SNR/sum((float(SNR)/float(SNR+1))**(n-(i+1)) for i in range(1,n+1))

digital_error_probability = lambda n, SNR: 1.0/2.0*(1.0-(1.0-2*q_func(np.sqrt(10**(SNR/10.0))))**n)
analog_error_probability = lambda n,SNR: q_func(np.sqrt(SNR_rho_n(n, 10**(SNR/10.0))))
```

```python
plt.figure().set_size_inches(20, 20)
plt.style.use('ggplot')

"""
Creo 4x2 tiles para los sub-plots
    +----+----+
    |  1 |  5 |
    +----+----+
    |  9 | 13 |
    +----+----+
    | 17 | 21 |
    +----+----+
    | 25 | -- |
    +----+----+
"""
grid = gridspec.GridSpec(4, 2)

"""
Itero por cada tile, y voy agregando los sub-plots 
asociados a cada n (de 1 a 25, con pasos de a 4).
Grafico las curvas para los sistemas analógicos y digitales.
"""
SNR_range = np.arange(-5,30,1)
n = 1
for tile in grid:
    if (n>25): 
        break;
    ax = plt.subplot(tile)
    ax.plot(SNR_range,
                  map(lambda SNR: digital_error_probability(n, SNR), SNR_range),
                  label="digital")
    ax.plot(SNR_range,
                  map(lambda SNR: analog_error_probability(n, SNR), SNR_range),
                  label="analogico")
    ax.set_yscale('log')
    ax.set_title("Probabilidad de Error n = " + str(n))
    ax.set_xlim([SNR_range[0], SNR_range[-1]])
    ax.set_ylim([10**(-6), 10**(0)])
    ax.legend(loc='upper right')
    n += 4

plt.show()
```

### Ejercicio 4 - Simulación Monte Carlo de las probabilidades de error

> _1. Realice un simulación Monte Carlo de ambos sistemas y calcule la probabilidad de error promedio para $SNR \in [5, 25]$ dB y $n = 9$. Grafique las curvas y compare los resultados con los resultados teóricos. Indique la cantidad de realizaciones Monte Carlo utilizadas para conseguir una buena estimación._

#### Analógico

Para realizar la simulación analógica se tomo como valores de $G_{i}$ los obtenidos en el *ejercicio 1*:

$${G_{i}}^{2} = \frac{SNR}{(SNR+1)\;h^{2}}$$

$$G_{i} = \sqrt{\frac{SNR}{(SNR+1)}}\;\frac{1}{h}$$

Como se observa de los resultados obtenidos en el *ejercicio 2*, el error depende exclusivamente de SNR, por lo tanto se fijaron los valores de $\sigma^{2} = 1$ y $h = 1$.

De esta forma, sabiendo que:

$$SNR = h^{2}A^{2}\;/\sigma^{2}$$

Se pudo definir:

$$SNR = A^{2}$$

#### Digital

De la misma forma, se tomaron los valores de $\sigma^{2} = 1$ y $h = 1$ para la simulación analógica. Y por lo tanto, se tomó también $SNR = A^{2}$ para calcular $A$ de forma tal de tener el $SNR$ deseado.

```python
h = 1

def simulacion_analogica(X1, A, n, SNR):
    Xi = X1
    Yi = 0
    for i in range(n):
        # Represento los ruidos como una normal de media 0 y varianza 1
        Wi = np.random.normal(loc=0.0,scale=1.0)
        Yi = h*Xi + Wi
        Gi = np.sqrt(float(SNR)/float(SNR+1))*1/float(h)
        Xi = Yi*Gi
    
    # Decisión sobre el símbolo transmitido
    return Yi

def monte_carlo_analogica(n, SNR, simulaciones):
    negativos = 0
    A = np.sqrt(10**(float(SNR)/float(10)))
    for i in range(simulaciones):
        # Obtengo X1 como 2A*B - A donde B es una Bernoulli(1/2)
        X1 = 2*A*np.random.randint(low=0.0, high=2.0)-A
        if X1 != np.sign(simulacion_analogica(X1, A, n, 10**(float(SNR)/float(10))))*A:
            negativos += 1
    return float(negativos)/float(simulaciones)
        

simulaciones = 100000
n = 9
SNR_range = np.arange(5,25,1)

plt.plot(SNR_range, 
         map(lambda SNR: analog_error_probability(n, SNR), SNR_range),
         label="analogico")
plt.plot(SNR_range, 
         map(lambda SNR: monte_carlo_analogica(n, SNR, simulaciones), SNR_range), 
         label = "simulacion analogica")    
    
plt.yscale('log')
plt.title("Probabilidad de Error n = " + str(n) + "\n Simulacion Montecarlo - Analogico")
plt.xlim([SNR_range[0], SNR_range[-1]])
plt.ylim([10**(-6), 10**(0)])
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.show()
```

**Análisis del gráfico (simulación analógica):**

Como se puede observar en el gráfico, se logró una buena estimación del error para un repetidor analógico. Es importante destacar, que a medida que aumenta SNR disminuye la probabilidad de error (acercándose cada vez más a 0). Esto implica que para poder distinguir estas probabilidades de 0 en la simulación monte carlo se requieren más realizaciones. A medida que aumentamos la cantidad de realizaciones, se podría decir que aumenta la "resolución". Para este caso, con $10^{5}$ realizaciones se obtiene una buena aproximación, aunque se puede observar que los últimos valores no se los puede distinguir de 0 (para solucionar esto, con $10^{6}$ realizaciones debería alcanzar).


```python
h = 1

def simulacion_digital(X1, A, n):
    Xi = X1
    for i in range(n):
        # Represento los ruidos como una normal de media 0 y varianza 1
        Wi = np.random.normal(loc=0.0,scale=1.0)
        Yi = h*Xi + Wi
        # Decisión sobre el símbolo transmitido
        Xi = np.sign(Yi)*A 
    return Xi

def monte_carlo_digital(n, SNR, simulaciones):
    negativos = 0
    A = np.sqrt(10**(float(SNR)/float(10)))
    for i in range(simulaciones):
        # Obtengo X1 como 2A*B - A donde B es una Bernoulli(1/2)
        X1 = 2*A*np.random.randint(low=0.0, high=2.0)-A
        if X1 != simulacion_digital(X1, A, n):
            negativos += 1
    return float(negativos)/float(simulaciones)


simulaciones = 100000
n = 9
SNR_range = np.arange(5,25,1)

plt.plot(SNR_range,
         map(lambda SNR: digital_error_probability(n, SNR), SNR_range),
         label="digital")
plt.plot(SNR_range, 
         map(lambda SNR: monte_carlo_digital(n, SNR, simulaciones), SNR_range), 
         label = "simulacion digital" )    
    
plt.yscale('log')
plt.title("Probabilidad de Error n = 9 \n Simulacion Montecarlo - Digital")
plt.xlim([SNR_range[0], SNR_range[-1]])
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.show()
```

**Análisis del gráfico (simulación digital):**

Como se puede observar en el gráfico, se podría decir que se logró una buena estimación del error de un repetidor digital para los primeros valores de $SNR$. Al igual que en el caso anterior, a medida que aumenta $SNR$ disminuye la probabilidad de error y se requieren más realizaciones para obtener una buena aproximación. En la simulación realizada se emplearon $10^{5}$ realizaciones, pero no fue suficiente en relación al orden de la probabilidad de error. 

Analizando en detalle los resultados de los dos gráficos, con $10^{5}$ realizaciones se obtiene una buena estimación hasta un orden de probabilidad de $10^{-5}$ aproximadamente. Por lo tanto, se podría decir que para mejorar la precisión se deberían utilizar aproximadamente $10^{13}$ realizaciones (debido a que las probabilidades en el gráfico alcanzan este orden aproximadamente).

El método de Monte Carlo funciona a partir de una secuencia de variables aleatorias independientes, las cuales son representadas por cada una de las simulaciones. 

$$\{X_{i}\}^{n}_{i=1}$$

Si se define una nueva variable aleatoria tal que:

$$Y_{i}(X_{1}) = \mathbb{1}\{X_{i} \neq X_{1}\}$$

En base a este resultado se puede estimar el valor de $\mathbb{E}[Y_{i}]$ como:

$$\mathbb{E}[Y_{i}] = \frac{1}{n}\sum^{n}_{i=1}Y_{i}(X1)$$

Y por la ley de los grandes numeros, este valor coincide con la probabilidad de error dado X1.

Como se puede ver, si conocemos $X_{1}$, el primer término pasa a representar la media de la normal $Y_{n}$. En cuanto al segundo término, como todas las $W_{i}$ son normales independientes (todas de media nula), la varianza de la suma de las normales será la suma de las varianzas (teniendo en cuenta el factor por el cual multiplica cada una).

Para ambos gráficos la probabilidad de error está representada por el area bajo la curva de los valores mayores o menores a 0 (según el valor de $X_{1}$) de los gráficos.

```python
# Distribución dado X1 = A

from scipy.stats import norm

n = 9
simulaciones = 100000

SNR = 5
A = np.sqrt(10**(float(SNR)/float(10)))

muestras_A = np.fromiter((simulacion_analogica(A, A, n, SNR) for i in range(simulaciones)), np.float )
plt.hist(muestras_A, bins = 100, alpha = 0.75, normed = True, label = "histograma | X1 = A")

muestras_nA = np.fromiter((simulacion_analogica(-A, A, n, SNR) for i in range(simulaciones)), np.float )
plt.hist(muestras_nA, bins = 100, alpha = 0.75, normed = True, label = "histograma | X1 = -A")

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlim([-10.0,10.0])

scale = float(sum(np.sqrt(float(SNR)/float(SNR+1))**(0.5*(n-(i+1))) for i in range(1,n+1)))

pdfs_A = lambda x, SNR : (norm.pdf((x-np.sqrt(float(SNR)/float(SNR+1))**(n-2)*A)/scale)/scale)
pdfs_nA = lambda x, SNR : (norm.pdf((x+np.sqrt(float(SNR)/float(SNR+1))**(n-2)*A)/scale)/scale)

x_range = np.arange(-10,10,0.01)

plt.plot(x_range, map(lambda x: pdfs_A(x,10**(SNR/10.0)), x_range), label = "pdfs | X1 = A")
plt.plot(x_range, map(lambda x: pdfs_nA(x,10**(SNR/10.0)), x_range), label = "pdfs | X1 = -A")

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlim([-10.0,10.0])
plt.show()
```
